```python
import pandas as pd;
import numpy as np;

file_path = '/content/database_24_25 (1).csv'
nba_data = pd.read_csv(file_path)
```

```python
print("First 5 row of the dataset:")
print(nba_data.head(10));
```

```python
#Display info on dataypes
nba_data.info()
```

```python
#find missing values
print('Print Missing Values in Each Column:');
print(nba_data.isnull().sum())
```

```python
#
print('Display the dataset to Help Detect Errors:')
print(nba_data.describe())
```

```python
#Only include players with minutes played over 0
nba_data = nba_data[nba_data['MP'] > 0]

#Check for missing values after preprocessing
print('Print Missing Values in Each Column:');
print(nba_data.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Points (PTS)
plt.figure(figsize=(10, 6))
sns.histplot(nba_data['PTS'], bins=30, kde=True)
plt.title('Distribution of Points (PTS)')
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Minutes Played (MP)
plt.figure(figsize=(10, 6))
sns.histplot(nba_data['MP'], bins=30, kde=True)
plt.title('Distribution of Minutes Played (MP)')
plt.xlabel('Minutes Played')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Assists (AST)
plt.figure(figsize=(10, 6))
sns.histplot(nba_data['AST'], bins=30, kde=True)
plt.title('Distribution of Assists (AST)')
plt.xlabel('Assists')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Rebounds (TRB)
plt.figure(figsize=(10, 6))
sns.histplot(nba_data['TRB'], bins=30, kde=True)
plt.title('Distribution of Total Rebounds (TRB)')
plt.xlabel('Rebounds')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Turnovers (TOV)
plt.figure(figsize=(10, 6))
sns.histplot(nba_data['TOV'], bins=30, kde=True)
plt.title('Distribution of Turnovers (TOV)')
plt.xlabel('Turnovers')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Game Score (GmSc)
plt.figure(figsize=(10, 6))
sns.histplot(nba_data['GmSc'], bins=30, kde=True)
plt.title('Distribution of Game Score (GmSc)')
plt.xlabel('Game Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
